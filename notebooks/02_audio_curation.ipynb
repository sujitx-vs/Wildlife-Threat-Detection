# Audio Exploration and Curation

```python
# imports
import pandas as pd
import numpy as np
```

Loading the processed metadata

```python
df = pd.read_csv('../datasets/processed/processed_metadata.csv')
```

```python
df.head()
```

Copying audio files into respective folders.

```python
import os
import shutil

SOURCE = "../datasets/raw/audio/"
DESTINATION = "../datasets/processed"
WILDLIFE = "/wildlife"
THREAT = "/threat"
WILDLIFE_PATH = DESTINATION + WILDLIFE
THREAT_PATH = DESTINATION + THREAT

cnts = os.listdir(SOURCE)

# folders are created before copying
os.makedirs(WILDLIFE_PATH, exist_ok=True)
os.makedirs(THREAT_PATH, exist_ok=True)

# Get a list (or set) of all filenames that belong to threat (label 0)
# This converts the Pandas Series into something Python can easily check against
threat_filenames = set(df['filename'][df['label'] == 1])
wildlife_filenames = set(df['filename'][df['label']== 0])

for filename in cnts:
    if filename in threat_filenames:
        #shutil is used to copy the files 
        shutil.copy(SOURCE + filename, THREAT_PATH)
    elif filename in wildlife_filenames:
        shutil.copy(SOURCE+filename,WILDLIFE_PATH)
```

Checking the number is matched or not

```python
wildlife_folder = os.listdir(WILDLIFE_PATH)
threat_folder = os.listdir(THREAT_PATH)

print("No. of audios in wildlife : ",len(wildlife_folder))
print("No. of audios in threat : ",len(threat_folder))

df['label'].value_counts()
```
